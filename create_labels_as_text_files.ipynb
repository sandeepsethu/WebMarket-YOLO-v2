#Creating text files for the labels/annotations of the bounding boxes


```python
import pandas as pd
import os
```

Reading the bounding box csv file

```python
df = pd.read_csv('db1_data.csv')
```

```python
df
```

Converting into the format necessary for YOLO with x and y center and width and height and setting class as 0

```python
img_h, img_w = (1704, 2272)
df.rename(columns={'filename':'image_id'}, inplace=True)
df['image_id'] = df['image_id'].apply(lambda x: x.split('.')[0])
df['x_center'] = (df['xmin'] + df['xmax'])/2
df['y_center'] = (df['ymin'] + df['ymax'])/2
df['w'] = df['xmax'] - df['xmin']
df['h'] = df['ymax'] - df['ymin']
df['classes'] = 0
df['x_center'] = df['x_center']/img_w
df['w'] = df['w']/img_w
df['y_center'] = df['y_center']/img_h
df['h'] = df['h']/img_h
df.head()

df=df.drop(['xmin', 'xmax', 'ymin', 'ymax', 'class','width','height'], axis=1)
```

```python
df.head()
```

```python
# Create an empty list
Row_bbox =[]
  
# Iterate over each row
for index, rows in df.iterrows():
    # Create list for the current row
    my_list =[rows.classes, rows.x_center, rows.y_center,rows.w, rows.h]
      
    # append the list to the final list
    Row_bbox.append(my_list)
```

```python
df["bbox"]= Row_bbox
```

```python
df
```

Writing the annotations into text files 

```python
unique_img_ids = df.image_id.unique()
```

```python
if not os.path.exists("../data/dataset/labels"):
    os.makedirs("../data/dataset/labels")
```

```python
folder_location = "../data/dataset/labels"
#change  unique_img_ids[:2] to unique_img_ids to iterate through all images
for img_id in unique_img_ids: # loop through all unique image ids. Remove the slice to do all images
    print(img_id)
    filt_df = df.query("image_id == @img_id") # filter the df to a specific id
    #print(filt_df.shape[0])
    all_boxes = filt_df.bbox.values
    file_name = "{}/{}.txt".format(folder_location,img_id) # specify the name of the folder and get a file name

    s = "%s %s %s %s %s \n" # the first number is the identifier of the class. If you are doing multi-class, make sure to change that
    with open(file_name, 'a') as file: # append lines to file
        for i in all_boxes:
            new_line = (s % tuple(i))
            file.write(new_line)  
```
